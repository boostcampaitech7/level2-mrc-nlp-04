## answer_start가 정확한가?

```python
# 정답 검증 함수
def verify_answer(row):
    context = row["context"]
    answer_text = row["answers"]["text"][0]
    answer_start = row["answers"]["answer_start"][0]

    # answer_start부터 answer_text 길이만큼의 문장을 추출
    extracted_answer = context[answer_start : answer_start + len(answer_text)]

    # 추출한 정답과 answer_text가 동일한지 검증
    return extracted_answer == answer_text


# 잘못된 정답을 포함한 행만 모아서 새로운 DataFrame 반환
def get_incorrect_rows(df):
    # 각 행에 대해 검증 수행하여 False인 행만 필터링
    incorrect_df = df[~df.apply(verify_answer, axis=1)]
    return incorrect_df
```

## Rule-base 에러 탐지

```python
t = train_df[["document_id", "question", "answers"]].copy()
t["answers"] = t["answers"].map(lambda x: x["text"][0])

v = valid_df[["document_id", "question", "answers"]].copy()
v["answers"] = v["answers"].map(lambda x: x["text"][0])
```

```python
from konlpy.tag import Kkma
import re
import pandas as pd

# Kkma 형태소 분석기 사용
kkma = Kkma()


def check_answer(df):
    wrong_data = []
    keywords = ["언제", "날짜", "누구", "어디", "무엇", "왜", "어떻게"]

    # DataFrame에서 각 행에 대해 검사
    for index, row in df.iterrows():
        question = row["question"]
        answer = row["answers"]

        # 질문에서 키워드 추출
        keyword = None
        tokens = kkma.pos(question)
        for word, tag in tokens:
            if word in keywords:
                keyword = word
                break

        if keyword is None:
            continue  # 질문에서 키워드가 발견되지 않으면 넘어감

        # 답변 유형 판단
        if keyword in ["언제", "날짜"]:
            # 숫자 있는지 체크
            if not bool(re.search(r"\d", answer)):
                wrong_data.append(row)
        elif keyword == "누구":
            # 사람 이름인지 판단 (KoNLPy를 사용하여 고유명사 추출)
            answer_tokens = kkma.pos(answer)
            person_found = False
            for word, tag in answer_tokens:
                if tag == "NNP" or tag == "NNG":  # 고유명사 또는 일반명사
                    person_found = True
                    break

            if not person_found:
                wrong_data.append(row)

    return pd.DataFrame(wrong_data, columns=df.columns)

def check_date_answer(df):
    def contains_keyword(text, keywords):
        for keyword in keywords:
            if keyword in text:
                return True
        return False

    wrong_data = []
    date_keywords = ["언제","날짜","해","년","월","일","때","시기","연도","기간"]

    # DataFrame에서 각 행에 대해 검사
    for index, row in df.iterrows():
        question = row["question"]
        answer = row["answers"]
        if contains_keyword(question,date_keywords):
            continue
        if answer.isdigit() or contains_keyword(answer,date_keywords):
            wrong_data.append(row)
    return pd.DataFrame(wrong_data, columns=df.columns)
```

## 정답 문장 추출

```python
# 정답이 포함된 문장을 추출하는 함수
def extract_answer_sentence(row):
    context = row["context"]
    answer_text = row["answers"]["text"][0]
    answer_start = row["answers"]["answer_start"][0]

    # 정답의 끝 위치 계산
    answer_end = answer_start + len(answer_text)

    # 문장의 시작 부분 찾기 (answer_start부터 왼쪽으로)
    sentence_start = answer_start
    while sentence_start > 0 and context[sentence_start - 1] not in [
        ".",
        "?",
        "!",
        "\n",
    ]:  # 문장 부호 및 개행 확인
        sentence_start -= 1

    # 문장의 끝 부분 찾기 (answer_end부터 오른쪽으로)
    sentence_end = answer_end
    while sentence_end < len(context) and context[sentence_end] not in [
        ".",
        "?",
        "!",
        "\n",
    ]:  # 문장 부호 및 개행 확인
        sentence_end += 1

    # 완전한 문장을 추출
    extracted_sentence = context[sentence_start:sentence_end].strip()

    return extracted_sentence
```

```python
valid_df["answer_sentence"] = valid_df.apply(extract_answer_sentence, axis=1)
```

```python
train_df["answer_sentence"] = train_df.apply(extract_answer_sentence, axis=1)
```
